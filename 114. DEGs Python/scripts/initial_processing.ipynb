```python
# import pandas to deal with data
import pandas as pd
```

```python
# read input files
X = pd.read_csv("VarX_Matrix.csv")
Y = pd.read_csv("VarY_Matrix.csv")
leafX = pd.read_csv("Leaf_DEGs_VarX_T1.csv")
leafY = pd.read_csv("Leaf_DEGs_VarY_T1.csv")
```

## Step 1
__Create all_VarX_TwoTimePoints.csv and all_VarY_TwoTimePoints.csv__

```python
# create matrices
df1 = pd.concat([X[['gene_name']], X[['VarXCRep.1', 'VarXCRep.2', 'VarXCRep.3']], X[['VarX1Rep.1', 'VarX1Rep.2', 'VarX1Rep.3']]], axis=1)
df2 = pd.concat([Y[['gene_name']], Y[['VarYCRep.1', 'VarYCRep.2', 'VarYCRep.3']], Y[['VarY1Rep.1', 'VarY1Rep.2', 'VarY1Rep.3']]], axis=1)
```

```python
# write them to a csv file
df1.to_csv("all_VarX_TwoTimePoints.csv", index=False)
df2.to_csv("all_VarY_TwoTimePoints.csv", index=False)
```

## Step 2
__Create Leaf_DEGs_VarX.csv and Leaf_DEGs_VarY.csv__

```python
# create dict from dataframe to access gene_name easily
X_dict = {df1.iloc[i, 0] : df1.iloc[i, 1:] for i in range(df1.shape[0])}
Y_dict = {df2.iloc[i, 0] : df2.iloc[i, 1:] for i in range(df2.shape[0])}
```

```python
# take subset from the differentially expressed genes data
leafX = leafX[['gene', 'log2FoldChange', 'padj', 'Athaliana_geneID', 'Gene_Function']]
leafY = leafY[['gene', 'log2FoldChange', 'padj', 'Athaliana_geneID', 'Gene_Function']]
```

```python
# add additional six columns for the expression of the three replicates
leaf_varx = pd.concat([leafX, pd.DataFrame([X_dict[leafX.iloc[i, 0]] for i in range(leafX.shape[0])]).reset_index(drop=True)], axis=1)
leaf_vary = pd.concat([leafY, pd.DataFrame([Y_dict[leafY.iloc[i, 0]] for i in range(leafY.shape[0])]).reset_index(drop=True)], axis=1)
```

```python
# write them into csv file
leaf_varx.to_csv("Leaf_DEGs_VarX.csv", index=False)
leaf_vary.to_csv("Leaf_DEGs_VarY.csv", index=False)
```

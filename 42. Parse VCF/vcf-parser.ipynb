```python
 def vcf_parser(filename):   
    with open(filename, "r") as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            if line.strip().startswith("#CHROM"):
                chrom_index = i
                break         
    return pd.read_csv(io.StringIO(''.join(lines[chrom_index:])), dtype={'#CHROM': str, 'POS': int, 'ID': str, \
        'REF': str, 'ALT': str, 'QUAL': str, 'FILTER': str, 'INFO': str}, sep='\t').rename(columns={'#CHROM': 'CHROM'})
```

```python
start_time = time.time()
df = vcf_parser(filename)
print("--- %s seconds ---" % (time.time() - start_time)) 
df.head()
```

```python
def read_vcf(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not l.startswith('##')]
    return pd.read_csv(
        io.StringIO(''.join(lines)),
        dtype={'#CHROM': str, 'POS': int, 'ID': str, 'REF': str, 'ALT': str,
               'QUAL': str, 'FILTER': str, 'INFO': str},
        sep='\t'
    ).rename(columns={'#CHROM': 'CHROM'})
```

```python
start_time = time.time()
df = read_vcf(filename)
print("--- %s seconds ---" % (time.time() - start_time)) 
df.head()
```

```python
df.info()
```

```python
def get_vcf_info(vcf_path):
    with open(vcf_path, "r") as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            if line.strip().startswith("#CHROM"):
                chrom_index = i
                break
        header = lines[chrom_index].strip().split("\t")

    return pd.read_csv(vcf_path, comment='#', delim_whitespace=True, header=None, names=header)
```

```python
start_time = time.time()
df = get_vcf_info(filename)
print("--- %s seconds ---" % (time.time() - start_time)) 
df.head()
```

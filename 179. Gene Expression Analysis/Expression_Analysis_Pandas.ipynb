```python
# Processing tabular data
import pandas as pd
```

```python
# Specify input filepath
varX_matrix_path = "Matrix/VarX_Matrix.csv"
varY_matrix_path = "Matrix/VarY_Matrix.csv"
varX_deg_path = "Matrix/Leaf_DEGs_VarX_T1.csv"
varY_deg_path = "Matrix/Leaf_DEGs_VarY_T1.csv"
```

```python
# Read Variety X matrix file
matrix_X = pd.read_csv(varX_matrix_path)
matrix_X.head()
```

```python
# Extract control condition columns
control_cols_X = sorted([col for col in matrix_X.columns if col.startswith("VarXC")])

# Extract stress treatment 1 condition columns
stress_treat_1_cols_X = sorted([col for col in matrix_X.columns if col.startswith("VarX1")])

print(f"Control Condition columns are: {control_cols_X}")
print(f"Stress Treatment Condition 1 columns are: {stress_treat_1_cols_X}")

# Combine all columns
varX_cols = ['gene_name'] + control_cols_X + stress_treat_1_cols_X
# Extract all selected columns from matrix_X and save to file
all_varX_df = matrix_X[varX_cols]
all_varX_df.to_csv("Matrix/all_VarX_TwoTimePoints.csv", index=False)
all_varX_df.head()
```

```python
# Read Variety Y matrix file
matrix_Y = pd.read_csv(varY_matrix_path)
matrix_Y.head()
```

```python
# Extract control condition columns
control_cols_Y = sorted([col for col in matrix_Y.columns if col.startswith("VarYC")])

# Extract stress treatment 1 condition columns
stress_treat_1_cols_Y = sorted([col for col in matrix_Y.columns if col.startswith("VarY1")])

print(f"Control Condition columns are: {control_cols_Y}")
print(f"Stress Treatment Condition 1 columns are: {stress_treat_1_cols_Y}")

# Combine all columns
varY_cols = ['gene_name'] + control_cols_Y + stress_treat_1_cols_Y
# Extract all selected columns from matrix_X and save to file
all_varY_df = matrix_Y[varY_cols]
all_varY_df.to_csv("Matrix/all_VarY_TwoTimePoints.csv", index=False)
all_varY_df.head()
```

```python
# Read Variety X DEGs file
deg_X = pd.read_csv(varX_deg_path).rename(columns={'gene': 'gene_name'})
deg_X.head()
```

```python
# List of desired columns
required_columns = ['gene_name', 'log2FoldChange', 'padj', 'Athaliana_geneID', 'Gene_Function']

# Extract desired columns
deg_X = deg_X[required_columns]
# Merge DEGs data with expression data on column gene_name and save to file
leaf_varX_df = deg_X.merge(all_varX_df, on='gene_name', how='inner')
leaf_varX_df.to_csv("Matrix/Leaf_DEGs_VarX.csv", index=False)
leaf_varX_df.head()
```

```python
# Read Variety Y DEGs file
deg_Y = pd.read_csv(varY_deg_path).rename(columns={'gene': 'gene_name'})
deg_Y.head()
```

```python
# List of desired columns
required_columns = ['gene_name', 'log2FoldChange', 'padj', 'Athaliana_geneID', 'Gene_Function']

# Extract desired columns
deg_Y = deg_Y[required_columns]
# Merge DEGs data with expression data on column gene_name and save to file
leaf_varY_df = deg_Y.merge(all_varY_df, on='gene_name', how='inner')
leaf_varY_df.to_csv("Matrix/Leaf_DEGs_VarY.csv", index=False)
leaf_varY_df.head()
```

```python
import os
import gzip
from Bio import PDB
from Bio.PDB.DSSP import DSSP
import numpy as np

def extract_essential_features(pdb_gz_path):
    """
    Extract essential features from a gzipped PDB file using DSSP for secondary structure.
    
    Returns:
        dict: Dictionary with key features and secondary structure sequence
    """
    parser = PDB.PDBParser(QUIET=True)
    
    # Create temporary unzipped file
    temp_pdb_path = pdb_gz_path.replace('.gz', '')
    with gzip.open(pdb_gz_path, 'rt') as f_in:
        with open(temp_pdb_path, 'w') as f_out:
            f_out.write(f_in.read())
    
    try:
        structure_id = os.path.basename(pdb_gz_path).split('.')[0]
        structure = parser.get_structure(structure_id, temp_pdb_path)
        model = structure[0]
        
        # Initialize feature dictionary
        features = {}
        features['PDB ID'] = structure_id
        features['n_residues'] = sum(1 for _ in model.get_residues())
        features['n_atoms'] = sum(1 for _ in model.get_atoms())
        
        # Calculate radius of gyration
        coords = np.array([atom.get_coord() for atom in model.get_atoms()])
        features['radius_of_gyration'] = np.sqrt(np.mean(np.sum((coords - np.mean(coords, axis=0))**2, axis=1)))

        # Get secondary structure using DSSP
        try:
            dssp = DSSP(model, temp_pdb_path, dssp='mkdssp')
            
            # Convert DSSP output to simple three-state classification
            ss_sequence = []
            phi_angles = []
            psi_angles = []
            h_bond_count = []  # Count of hydrogen bonds per residue

            for key in dssp.keys():
                ss = dssp[key][2]
                if ss in ['H', 'G', 'I']:  # Alpha helix and similar
                    ss_sequence.append('H')
                elif ss in ['B', 'E']:     # Beta sheet
                    ss_sequence.append('E')
                else:                      # Coil and everything else
                    ss_sequence.append('C')

                # Angles
                phi_angles.append(dssp[key][4])
                psi_angles.append(dssp[key][5])
                
                # Count hydrogen bonds for this residue
                h_bonds = 0
                if dssp[key][6] != 0: h_bonds += 1  # NH->O_1
                if dssp[key][8] != 0: h_bonds += 1  # O->NH_1
                if dssp[key][10] != 0: h_bonds += 1 # NH->O_2
                if dssp[key][12] != 0: h_bonds += 1 # O->NH_2
                h_bond_count.append(h_bonds)

            features['ss_sequence'] = ''.join(ss_sequence)
            features['h_bond_density'] = np.mean(h_bond_count)
            features['phi_psi_ratio'] = np.mean(np.abs(phi_angles)) / np.mean(np.abs(psi_angles))
    
            # Calculate percentages
            total_ss = len(ss_sequence)
            if total_ss > 0:
                features['helix_percent'] = 100 * ss_sequence.count('H') / total_ss
                features['sheet_percent'] = 100 * ss_sequence.count('E') / total_ss
                features['coil_percent'] = 100 * ss_sequence.count('C') / total_ss
            
        except Exception as e:
            print(f"DSSP calculation failed for {structure_id}: {e}")
            features['ss_sequence'] = 'X' * features['n_residues']
            features['helix_percent'] = features['sheet_percent'] = features['coil_percent'] = float('nan')
            features['avg_accessibility'] = features['max_accessibility'] = float('nan')
        
        return features
        
    finally:
        if os.path.exists(temp_pdb_path):
            os.remove(temp_pdb_path)

def process_directory(directory_path):
    """Process all PDB files in a directory"""
    all_features = []
    
    for filename in os.listdir(directory_path):
        if filename.endswith('.pdb.gz'):
            try:
                features = extract_essential_features(os.path.join(directory_path, filename))
                all_features.append(features)
                print(f"Processed {filename}")
            except Exception as e:
                print(f"Error processing {filename}: {e}")
    
    return all_features

# Example usage:
#features_list = process_directory('structures')
#struct_feat_df = pd.DataFrame(features_list)
#struct_feat_df
```
